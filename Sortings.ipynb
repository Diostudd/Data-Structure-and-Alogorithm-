```python
#Now Use Bubble Sorting to Sort these elements


def bubblesorting(elements):
    size = len(elements)      #To find the Size and how many times the loop should run

    #Now Running the first Loop
    for j in range(size-1): #why do we run loop n-1 times
        for i in range(size-1):#why do we run loop n-1 times
            if elements[i] > elements[i+1]:
                tem = elements[i]
                elements[i] = elements[i+1]
                elements[i+1] = tem

if __name__ == "__main__":#why do we write this?

    a = [12,32,42,121,12,12,3,4,5,67]
    bubblesorting(a)#Why cant we do print(bubblesoting(a))
    print(a)

    b = [12,1,1,1,1,2,2,2,3,3,1,212,1,2,1312,23,13,2,23,12,312,3]
    bubblesorting(b)
    print(b)
```
